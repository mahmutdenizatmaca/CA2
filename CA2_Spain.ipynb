```python
# Importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt   
import seaborn as sns
%matplotlib inline
```

```python
# deleting unnecessary columns
data_spa = data_spa.drop(['POTENCIA', 'CILINDRADA',
       'COMBUSTIBLE_DGT', 'TIPUSALIMENTACIÓ_DGT', 'NIVELLEMISSIONS_DGT',
       'CATELLECTRICA_DGT', 'AUTONOMIAELECTRICA_DGT', 'CLASAMBIENTAL_DGT',
       'BONIFICACIO', 'BONIFICACIO_CODI', 'BONIFICACIO_DESC',
       'TITULAR_PERSONALITAT','EXTRACCIO', 'TITULAR_CRIPTODOC',
       'TITULAR_CRIPTONIA', 'X', 'Y'] ,axis=1)
```

```python
# changing column names
col_names = ["Type of Vehicle","Year","Value"]
data_spa.columns = col_names
col_names
```

```python
# encoding
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data_spa["Type of Vehicle"] = le.fit_transform(data_spa["Type of Vehicle"])
```

```python
# Spliting the dataset into independent and dependent variables

X = data_spa.iloc[:,0:1].values  
y = data_spa.iloc[:,-1].values 
print(X.shape)
print(type(X))
```

## linear regression

```python
# Using linear regression
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

```python
nan_indices = np.isnan(y_train)
print(np.where(nan_indices))
```

```python
from sklearn.impute import SimpleImputer

# assigning a mean value to missing data
imputer = SimpleImputer(strategy='mean')
y_train = imputer.fit_transform(y_train.reshape(-1, 1)).flatten()
```

```python
from sklearn.linear_model import LinearRegression
modelRegresyon = LinearRegression()
modelRegresyon.fit(X_train, y_train)
```

```python
y_pred = modelRegresyon.predict(X_test)
```

```python
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, modelRegresyon.predict(X_train), color = 'blue')
plt.show()
```

```python
plt.scatter(X_test, y_test, color = 'green')
plt.plot(X_test, modelRegresyon.predict(X_test), color = 'blue')
plt.show()
```
